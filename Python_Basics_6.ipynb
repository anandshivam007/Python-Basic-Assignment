```python
1. What are escape characters, and how do you use them?

Escape characters are used to  insert characters that are illegal in a string. 
we can use it with Backslash \ followed by the character you want to insert in the string 
e.g - inserting a double quotes inside a single quote a = 'I am Anand \"Shivam\" from Bangalore.'
```

```python
2. What do the escape characters n and t stand for?

\n - new line
\t - tab
```

```python
 3. What is the way to include backslash characters in a string?

    We can Include it by double backslash inside quotes
    
    e.g - print('I am Anand "\\"Shivam  from Bangalore.')
```
